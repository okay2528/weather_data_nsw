```python
import pandas as pd
import numpy as np
import os
```

```python
## change directory and get list of file
os.chdir("c:\\Users\\Windows\\Documents\\2.Project\\5.Weather\\2.Bronze")

filelist = os.listdir()
```

```python
## read files
df = pd.DataFrame()
for i in filelist:
    print(i)
    temp = pd.read_csv(i, encoding='utf-8', index_col = 0)
    temp = temp[temp['station_name']!= 'Totals:']
    df = pd.concat([df, temp], ignore_index=True)
    del temp

filepath = f'C:\\Users\\Windows\\Documents\\2.Project\\5.Weather\\3.Silver\\NSW_weather_data_12mth.csv'
        
df.to_csv(filepath, sep=',', encoding='utf-8')
```
